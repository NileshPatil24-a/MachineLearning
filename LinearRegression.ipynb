```python
import pandas  as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day48-simple-linear-regression/placement.csv")
```

```python
df.tail()
```

```python
plt.figure(figsize=(3,3))
sns.scatterplot(x = df['cgpa'], y=df['package'])
plt.show()
```

```python
x =df.iloc[:,0:1]
y = df.iloc[:,1:2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
lr.predict(x_test)
```

```python
plt.figure(figsize=(3,3))
plt.scatter(df['cgpa'],y=df['package'])
plt.plot(x_test, lr.predict(x_test), color = "red")
plt.xlabel("cgpa")
plt.ylabel('package')
plt.show()
```
