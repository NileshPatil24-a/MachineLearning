```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 

warnings.filterwarnings("ignore")
```

```python
df = sns.load_dataset("titanic")
```

```python
df_required = df[["sex","class","who","alive","alone"]]
```

```python
df_required.sample(5)
```

```python
for i in df_required.columns:
    print(df_required[i].value_counts())
    print()
```

```python
df_required["sex"] = df_required["sex"].map({"male":0,"female":1})
```

```python
df_required["class"] = df_required["class"].map({"First":0,"Second":1, "Third" : 2})
```

```python
df_required["alive"] = df_required["alive"].map({"yes":1,"no":0})
```

```python
df_required.head(5)
```

```python
pd.get_dummies(df_required, columns=["who"] , drop_first=True)
```

```python
df_required.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first')
```

```python
trans_form = ohe.fit_transform(df_required[["who"]]).toarray()
```
